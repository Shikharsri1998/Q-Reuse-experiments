```python
import json
import matplotlib.pyplot as plt

with open("Fidelity_scripts_comparison.json", "r") as jsonfile:
    data = json.load(jsonfile)
```

```python
data
```

```python
categories = ["Base Simulator", "Base Real", "Q Reuse Simulator", "Q Reuse Real"]

Fidelities=[]
Fidelities.append(data['Fidelity_baseline_simulator'])
Fidelities.append(data['Fidelity_baseline_real'])
Fidelities.append(data['Fidelity_qubit_reuse_simulator'])
Fidelities.append(data['Fidelity_qubit_reuse_real'])

bars = plt.bar(categories, Fidelities)

plt.xticks(rotation=90,fontsize=24)
plt.ylabel('Fidelity',labelpad=8,fontsize=24)

# Add y-axis major and minor grid lines
plt.grid(which='both', axis='y', linestyle='--', linewidth=0.8, color='gray')
plt.grid(which='minor', axis='y', linestyle=':', linewidth=0.5, color='gray')
plt.minorticks_on()
```
